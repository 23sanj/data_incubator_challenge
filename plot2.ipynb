```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

file1='/Users/sanjana/Desktop/project/crime/17_Crime_by_place_of_occurrence_2001_2012.csv'
df = pd.read_csv(file1)
#Selecting the major states by geographical region
df=df[(df['STATE/UT'] =='ANDHRA PRADESH') | (df['STATE/UT'] =='MAHARASHTRA') | (df['STATE/UT'] =='MADHYA PRADESH') | (df['STATE/UT'] =='RAJASTHAN') | (df['STATE/UT'] =='UTTAR PRADESH')]

```

```python
#Theft crime
crimeDf = df[['TOTAL - Theft','YEAR','STATE/UT']].groupby(['YEAR','STATE/UT']).sum()
ct=crimeDf.unstack(0).plot.barh()
#ct.set_xlabel('Area Name')
ct.set_ylabel('Theft  Crime Occurences per state from 2001 to 2012')
ct.set_title('Thefts occurence from  2001 to 2012')
plt.show()
```

```python
#Crime against women
file2='/Users/sanjana/Desktop/project/crime/42_District_wise_crimes_committed_against_women_2001_2012.csv'
df = pd.read_csv(file2)
df=df[(df['STATE/UT'] =='ANDHRA PRADESH') | (df['STATE/UT'] =='MAHARASHTRA') | (df['STATE/UT'] =='MADHYA PRADESH') | (df['STATE/UT'] =='RAJASTHAN') | (df['STATE/UT'] =='UTTAR PRADESH')]
crimeDf2=df[['Rape','Year','STATE/UT']].groupby(['Year','STATE/UT']).sum()
ct=crimeDf2.unstack(0).plot.barh()
ct.set_ylabel('Rape  Crime Occurences per state from 2001 to 2012')
ct.set_title('Rape occurence from  2001 to 2012')
plt.show()
```
