```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mplt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Crime against women
file1='/Users/sanjana/Desktop/project/crime/42_District_wise_crimes_committed_against_women_2001_2012.csv'
file2='/Users/sanjana/Desktop/project/crime/42_District_wise_crimes_committed_against_women_2013.csv'
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df =df2.append(df1, ignore_index=True)
#crimeDf2=df[['Rape','Year']].groupby(['Year']).sum()
#ct=crimeDf2.unstack(0).plot.barh()
#ct.set_ylabel('Rape  Crime Occurences from 2001 to 2012')
#ct.set_title('Rapes occurence from  2001 to 2012')
#plt.show()
df= df.fillna(0)
df =df.drop(['DISTRICT'], axis=1)

#How does it increase over the year?
crimeDf=df.groupby(['Year']).sum()
ct=crimeDf.plot.barh()
ct.set_ylabel('Number of Crime Incidents Against Women')
ct.set_title('Crime Incidents Against Women from 2001 to 2013')
mplt.pyplot.savefig('crime_women_rate.png')
plt.show()
crimeDf
```
